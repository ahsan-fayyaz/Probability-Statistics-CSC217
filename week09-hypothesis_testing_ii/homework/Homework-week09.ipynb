```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import math
```

```python
sv = pd.read_csv('/home/ahsan/Spring2019/csc217/Project/Temperature_data/Full_temp.csv',header=0)
sv
```

```python
mean = np.mean(sv['JUN'])
print("Mean =", mean)
```

```python
var = np.var(sv['JUN'])
print("Variance =", var)
```

```python
standard_dev = math.sqrt(var)
print("Standard deviation =", standard_dev)
```

### A PDF function for whole data

```python
dist = stats.norm(mean, standard_dev)
fig = plt.figure(figsize=(10,10))
x = np.linspace(dist.ppf(0.0001), dist.ppf(0.9999), 100)
plt.plot(x, dist.pdf(x))
plt.plot([dist.ppf(0.95), dist.ppf(0.95)],[0, dist.pdf(dist.ppf(0.95))], linestyle='dotted', color='#1f77b4')
X_fill = np.linspace(dist.ppf(0.95), dist.ppf(0.9999))
Y = dist.pdf(X_fill)
plt.fill_between(X_fill, 0, Y, facecolor='red')
plt.ylim(bottom=0)
plt.xlabel('Temperatures')
plt.ylabel('Probability Density Function')
fig.suptitle('Distribution of Temperatures in JUNE', fontsize=15, y=0.92)
```

### Z-test
Let's do a one sided Z-test here. From our data, we know that the mean temperatures of NYC in June are 71F.
#### Our NULL hypothesis: 
                          The temperatures in JUNE in NYC are 71F
#### Alternative hypothesis:
                          The temperatures in JUNE in NYC are greater than 75F
                        
At Significance level = 0.05
We will reject the null hypothesis that MEAN temperaturs in NYC in the month of JUNE are 71F

### Plot a PDF Graph with Mean 71 and standard deviation (based on the 20 temperatures)

```python
dist = stats.norm(mean, standard_dev/np.sqrt(20))
fig = plt.figure(figsize=(10,10))
x = np.linspace(dist.ppf(0.0001), dist.ppf(0.9999), 100)
plt.plot(x, dist.pdf(x))
plt.plot([dist.ppf(0.95), dist.ppf(0.95)],[0, dist.pdf(dist.ppf(0.95))], linestyle='dotted', color='#1f77b4')
X_fill = np.linspace(dist.ppf(0.95), dist.ppf(0.9999))
Y = dist.pdf(X_fill)
plt.fill_between(X_fill, 0, Y, facecolor='red')
plt.ylim(bottom=0)
plt.xlabel('Temperatures')
plt.ylabel('Probability Density Function')
fig.suptitle('Distribution of Temperatures in JUNE', fontsize=15, y=0.92)
```
